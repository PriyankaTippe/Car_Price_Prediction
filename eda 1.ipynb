```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv("new_cleaned_data.csv")
ds
```

```python
ds.reset_index(inplace=True)
```

```python
ds[["company","name","year","kms_driven","fuel_type","Price"]].to_csv("new_cleaned_data.csv")
```

```python
#barplot

companywisecount=ds.groupby("company")["company"].count()
plt.figure(figsize=(15,5))
ax=companywisecount.plot(kind='bar',color='purple',title='company-wise count')
plt.xticks(rotation=45)
for i,v in enumerate(companywisecount):
    ax.text(i,v+0.1,str(v),ha='center',va='bottom')
plt.xlabel=('company')
plt.ylabel=('count')
plt.show()
```

```python
# count of cars by year in barplot

vehicle_count=ds["year"].value_counts().sort_index()
plt.figure(figsize=(15,6))
vehicle_count.plot(kind='bar',color='blue')
plt.title('count of car by year')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
# pie chart


def autopct_with_count(pct,all_vals):
    absolute=int(round(pct/100.*sum(all_vals)))
    return f"{absolute}\n({pct:.1f}%)"
fuel_count=ds["fuel_type"].value_counts()
plt.figure(figsize=(5,5))
plt.pie(fuel_count.values,labels=fuel_count.index, autopct=lambda pct:autopct_with_count(pct,fuel_count.values),startangle=0,colors=plt.cm.Paired.colors)
plt.title('Proportion of cars by fuel type')
plt.tight_layout()
plt.show()
```

```python
# swarm plot        (It shows concentration of price for rach fuel type,  outlier shows high price of car)

plt.figure(figsize=(10,6))
sns.swarmplot(x='fuel_type',y='Price',data=ds,palette='Paired')
plt.title('Cars Prices by Fuel Type')
plt.tight_layout()
plt.show()
```

```python
# histogram for Price,year,kms_driven       (It shows how a numerical variables distributed)

plt.figure(figsize=(8,3))
plt.hist(ds['Price'],bins=20,color='purple',edgecolor='black')
plt.title('Distribution of car Prices')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,3))
plt.hist(ds['year'],bins=len(ds['year'].unique()),color='lightgreen',edgecolor='black')
plt.title('Distribution of car manufacturing years')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8,3))
plt.hist(ds['kms_driven'],bins=20,color='salmon',edgecolor='black')
plt.title('Distribution of kms_driven')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```
